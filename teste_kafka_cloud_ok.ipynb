```python
## importando os módulos

from confluent_kafka import Producer
from faker import Faker
import random
import datetime
import time
import json

fake = Faker()  ## faker utilizado para gerar nomes fakes
```

```python
# Definindo o producer, e as propriedades necessárias do cluster Confluent Cloud Kafka,
# como API KEY, Servidor, username...

# CLUSTER FOI EXCLUÍDO !!!

p = Producer({
"bootstrap.servers": "pkc-lzvrd.us-west4.gcp.confluent.cloud:9092",
"security.protocol": "SASL_SSL",
"sasl.mechanisms": "PLAIN",
"sasl.username": "AAAAAAAAAAAAAA",
"sasl.password": "4q/WIz0Avh2LJd0p9AAAAAAAAAAAAAAAAAwXcFPmLbwhiICl0VEKGb/5lT"
})

# definindo variável para pegar posição do datetime
current_time = datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S")


# criando função para definir status das mensagens
def def_status(err, msg):
   
    if err is not None:
        print('Mensagem não entregue erro : {}'.format(err))
    else:
        print('Mensagem enviada para o topic {} Partição   [{}]'.format(msg.topic(), msg.partition()))

#defininfo função para enviar as mensagens
def send_mensagens():
  for i in range(0,1000):              ## um loop para gerar 1000 mensagens
    msg= {"ID": random.randint(0,20),  ## gerando valores int randômicos entre 0 e 20
      "NOME":  (fake.name()),          ## gerando nomes fakes
      "DATE": current_time }           ## definindo horário de envio
    msg= json.dumps(msg)               ## transformando a mensagem para Json
    time.sleep(1)                      ## definindo para enviar a cada 1 segundo
    print(msg)
    
        
    p.produce('topic_guma', value=str(msg), callback=def_status)
    p.poll(0)
    

 # start 
if __name__ == '__main__':
    send_mensagens()
```
